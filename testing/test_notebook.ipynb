```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as anim
import plasmapy as plas
```

Code

```python
# Learning how to use matplotlib.animation, so here's a test animation :)

# function determining the animated curve
def f(x, mod):
    return np.cos(mod/2 + x) * (np.e**(-(x/mod)**2))/mod

# set range of points for x-axis and generate y-values from each x
x_data = np.arange(-3, 3, 0.01)
y_data = f(x_data, 1)

# func to update the plot for each new frame
def update(frame):
    y_data = f(x_data, 1 + 0.3 * frame)
    line.set_xdata(x_data)
    line.set_ydata(y_data)
    return (line)

fig, ax = plt.subplots()

line = ax.plot(x_data, y_data)[0]
ax.set_title("Test Animation")
ax.set_ylim(-0.5, 0.5)

# create/export anim
ani = anim.FuncAnimation(fig=fig, func=update, frames=60, interval=30)
ani.save("test_animation.gif")
```
