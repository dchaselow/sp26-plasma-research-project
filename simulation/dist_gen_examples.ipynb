```python
from dist_gen import *
import matplotlib.pyplot as plt
```

```python
n_samples = 500000

def dist_func(x):
    return x**2
rfd_test = rfd_array(dist_func, n_samples, -20, 20, 1200, True)

fig, ax = plt.subplots()
ax.hist(rfd_test, bins = 1000)
ax.set_title("$f(\zeta)=\zeta^2$, $N=500000$")
ax.set_box_aspect(1)
ax.set_xlabel("Parameter $(\zeta)$")
ax.set_ylabel("Count")
fig.savefig(r"outputs\dist_gen_examples\z_squared")
```

```python
n_samples = 500000

def dist_func_2(x):
    return np.e**(-x**2)
rfd_test_2 = evd_array(dist_func_2, n_samples, -20, 20, 1200, True)

fig, ax = plt.subplots()
ax.hist(rfd_test_2, bins = 1000)
ax.set_title("$f(\zeta)=e^{-\zeta^2}$, $N=500000$")
ax.set_box_aspect(1)
ax.set_xlabel("Parameter $(\zeta)$")
ax.set_ylabel("Count")
fig.savefig(r"outputs\dist_gen_examples\gaussian")
```

```python
n_samples = 500000

def dist_func_3_x(x):
    return 1

def dist_func_3_v(v):
    return np.e**(-(5*v-3)**2) + np.e**(-(5*v+3)**2)

rfd_test_3_x = evd_array(dist_func_3_x, n_samples, -20, 20, 1200)

rfd_test_3_v = evd_array(dist_func_3_v, n_samples, -20, 20, 1200)

fig, ax = plt.subplots()
ax.hist2d(rfd_test_3_x, rfd_test_3_v, bins = 250)
ax.set_title("Phase Space Distribution, $N=500000$")
ax.set_box_aspect(1)
ax.set_xlabel("Position (m)")
ax.set_ylabel("Velocity (m/s)")
fig.savefig(r"outputs\dist_gen_examples\phase_space")
```
